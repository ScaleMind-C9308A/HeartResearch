```python
import matplotlib.pyplot as plt
import scipy.io
from scipy import signal
import numpy as np
import pywt
import scaleogram as scg
```

```python
mat = scipy.io.loadmat('/Users/hoangminhthao/Downloads/A4472.mat')
```

```python
a = mat['ECG']
```

```python
b = a[0][0][2][0][0:400]
```

```python
type(b)
scale = np.arange(1,401)
```

```python
coef, freqs = pywt.cwt(b, scale, 'morl')
plt.figure(figsize=(150,150))
plt.imshow(np.array(coef), 
    extent=[0, 200, 30, 1], 
    interpolation = 'bilinear', cmap='jet')

```

```python
# choose default wavelet function 
scg.set_default_wavelet('morl')

nn = 33
signal_length = 128
# range of scales to perform the transform
scales = scg.periods2scales( np.arange(1, signal_length+1) )
x_values_wvt_arr = range(0,len(b[nn]),1)

# plot the signal 
fig1, ax1 = plt.subplots(1, 1, figsize=(9, 3.5));  
ax1.plot(x_values_wvt_arr, b[nn], linewidth=3, color='blue')
ax1.set_xlim(0, signal_length)
ax1.set_title("ECG ARR")

# the scaleogram
scg.cws(b[nn][:signal_length], scales=scales, figsize=(10, 4.0), coi = False, ylabel="Period", xlabel="Time",
        title='ECG_ARR: scaleogram with linear period'); 

print("Default wavelet function used to compute the transform:", scg.get_default_wavelet(), "(",
      pywt.ContinuousWavelet(scg.get_default_wavelet()).family_name, ")")
```
